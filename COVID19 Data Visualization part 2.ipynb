```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.dates as mdates

%matplotlib inline

plt.style.use('fivethirtyeight')
```

```python
df1 = pd.read_csv('https://raw.githubusercontent.com/datasets/covid-19/main/data/countries-aggregated.csv', parse_dates = ['Date'])
df1.head()
```

```python
df1['Total Cases'] = df1[['Confirmed','Recovered', 'Deaths']].sum(axis=1)
df1.head()
```

```python
worldwide_df = df1.groupby(['Date']).sum()
worldwide_df.head()
```

```python
plot1 = worldwide_df.plot(figsize=(16,10))
plot1.set_xlabel('Date')
plot1.set_ylabel('Number of Cases Worldwide')
plot1.title.set_text('Worldwide COVID Insights')

plt.show()
```

### United Kingdom Daily Cases and Deaths

```python
uk_df = df1[df1['Country'] =='United Kingdom'].groupby(['Date']).sum()

fig = plt.figure(figsize=(16,10))
ax = fig.add_subplot(111)

ax.plot(worldwide_df[['Total Cases']], label='Worldwide')
ax.plot(uk_df[['Total Cases']], label='UK')
ax.set_xlabel('Date')
ax.set_ylabel('Number of Cases Worldwide')
ax.title.set_text('Worldwide vs UK Total Cases')

plt.legend(loc='upper left')
plt.show()
```

```python
uk_df = uk_df.reset_index()
uk_df.head()
```

```python
uk_df['Daily Confirmed'] = uk_df['Confirmed'].sub(uk_df['Confirmed'].shift())
uk_df
```

```python
uk_df['Daily Deaths'] = uk_df['Deaths'].sub(uk_df['Deaths'].shift())
uk_df
```

```python
fig = plt.figure(figsize=(20,8))
ax = fig.add_subplot(111)

ax.bar(uk_df['Date'], uk_df['Daily Confirmed'], color='b', label='US Daily Confirmed Cases')
ax.bar(uk_df['Date'], uk_df['Daily Deaths'], color='r', label='US Daily Deaths')
ax.set_xlabel('Date')
ax.set_ylabel('Number of People Affected')
ax.title.set_text('UK Daily Cases and Deaths')

plt.legend(loc='upper left')
plt.show()
```
